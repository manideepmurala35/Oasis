```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = '/content/Unemployment in India.csv'
unemployment_data = pd.read_csv(file_path)
```

```python
print("First few rows of the dataset:")
print(unemployment_data.head())
```

```python
unemployment_data.columns = unemployment_data.columns.str.strip()
```

```python
unemployment_data['Date'] = pd.to_datetime(unemployment_data['Date'], dayfirst=True)
```

```python
print("\nSummary statistics of the estimated unemployment rate:")
print(unemployment_data['Estimated Unemployment Rate (%)'].describe())
```

```python
plt.figure(figsize=(10, 6))
plt.plot(unemployment_data['Date'], unemployment_data['Estimated Unemployment Rate (%)'], marker='o', linestyle='-', color='b')
plt.title('Estimated Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(unemployment_data['Estimated Unemployment Rate (%)'], bins=20, edgecolor='black')
plt.title('Distribution of Estimated Unemployment Rate')
plt.xlabel('Estimated Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

```python
unemployment_data['Month'] = unemployment_data['Date'].dt.to_period('M')
monthly_avg_unemployment = unemployment_data.groupby('Month')['Estimated Unemployment Rate (%)'].mean()
```

```python
plt.figure(figsize=(12, 6))
monthly_avg_unemployment.plot(kind='bar', color='skyblue')
plt.title('Monthly Average Estimated Unemployment Rate')
plt.xlabel('Month')
plt.ylabel('Average Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```
